```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import invert
import numpy as np
```

```python

retinaImage = data.retina()
horseImage = data.horse()

retinaCropped = retinaImage.copy()
retinaCropped = retinaImage[0:256,64:320]

horseCropped =horseImage.copy()
horseCropped = horseCropped[64:256,128:320]

fig, axes = plt.subplots(2, 2, figsize = (10, 10))
ax = axes.ravel()

ax[0].imshow(retinaImage)
ax[0].set_title("Citra Input 1")
ax[1].imshow(horseImage, cmap='gray')
ax[1].set_title("Citra Input 2")
ax[2].imshow(retinaCropped)
ax[2].set_title("Citra Output 1")
ax[3].imshow(horseCropped, cmap='gray')
ax[3].set_title("Citra Output 2")

fig.tight_layout
plt.show()
```
