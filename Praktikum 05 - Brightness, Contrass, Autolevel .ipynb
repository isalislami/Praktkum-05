```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
img = imageio.imread("apel2.jpg")
```

```python
img_width= img.shape[1]
img_channel= img.shape[2]
img_type = img.dtype
```

```python
#membuat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype=np.uint8)
```

```python
#membuat fungsi penambahan brightness dengan nilai yang menjadi parameter
def brighter(nilai):
    for y  in range(0, img_height):
        for x in range(0, img_width):
            red= img[y][x][0]
            green= img[y][x][1]
            blue= img[y][x][2]
            gray= (int(red)+ int(green)+int(blue))/3
            gray += nilai
            if gray > 255:
                gray = 255
            if gray<0:
                gray = 0
            img_brightness[y][x] = (gray, gray, gray)
```

```python
#menampilkan gambar dengan parameter -100 dan +100
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

brighter(-100)
ax[0].imshow(img_brightness) 
ax[0].set_title("Brightness -100") 

brighter(100)
ax[1].imshow(img_brightness) 
ax[1].set_title("Brightness +100") 

fig.tight_layout
plt.show()
```

```python
#Brightness RGB
img_rgbbright = np.zeros(img.shape, dtype=np.uint8)
```

```python
#fungsi untuk brightness RGB dengan nilai parameter 
def rgbbright(nilai):
    for y in range (0, img_height):
        for x in range (0, img_width):
            red= img[y][x][0]
            red+= nilai
            if red > 255:
                red = 255
            if red < 0:
                red = 0
            green= img[y][x][1]
            green+= nilai
            if green > 255:
                green = 255
            if green < 0:
                green = 0
            blue= img[y][x][2]
            blue+= nilai
            if blue > 255:
                blue = 255
            if blue < 0:
                blue = 0
            img_rgbbright[y][x] = (red, green, blue)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

rgbbright(-100)
ax[0].imshow(img_rgbbright) 
ax[0].set_title("Brightness -100") 

rgbbright(100)
ax[1].imshow(img_rgbbright) 
ax[1].set_title("Brightness +100") 
fig.tight_layout
plt.show()
```

```python
#contrast
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def contrass(nilai):
    for y in range (0, img_height):
        for x in range (0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray= (int(red)+ int(green)+int(blue))/3
            gray+= nilai
            if gray > 255:
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) 
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

contrass(2)
ax[0].imshow(img_contrass) 
ax[0].set_title("Contrass 2") 

contrass(200)
ax[1].imshow(img_contrass) 
ax[1].set_title("Contrass 3") 

fig.tight_layout
plt.show()
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()
```
